```python
from ipywidgets import Button, Layout, HBox
from IPython.display import display
import IPython

buttonDokuDaten = Button(description = 'Standortdaten einlesen', layout = Layout(width='50%'))
buttonStandort = Button(description = 'Standorte ausgeben', layout = Layout(width='50%'))

global zaehlerDaten
zaehlerDaten = 0

with open('Textdateien/Datenauswertung/DokumentationDaten.txt', 'r') as DateiDaten:
    datenDaten = DateiDaten.read()
    pass

def datenButtonGeklickt(b):
    global zaehlerDaten
    global zaehlerStandorte
    if((zaehlerStandorte + zaehlerDaten) % 2 == 0):
        IPython.display.clear_output(wait = False)
        display(HBox([buttonDokuDaten,buttonStandort]))
        display(IPython.display.Markdown(datenDaten))
        zaehlerDaten = zaehlerDaten + 1
        pass
    else:
        if(zaehlerDaten % 2 == 0):
            IPython.display.clear_output(wait = False)
            display(HBox([buttonDokuDaten,buttonStandort]))
            display(IPython.display.Markdown(datenDaten))
            zaehlerStandorte = zaehlerStandorte + 1
            zaehlerDaten = zaehlerDaten + 1
            pass
        else:
            IPython.display.clear_output(wait = False)
            display(HBox([buttonDokuDaten,buttonStandort]))
            zaehlerDaten = zaehlerDaten + 1
            pass
        pass
    pass

buttonDokuDaten.on_click(datenButtonGeklickt)

global zaehlerStandorte
zaehlerStandorte = 0

with open('Textdateien/Datenauswertung/StandorteAusgabe.txt', 'r') as DateiSandorte:
    standorteDaten = DateiSandorte.read()
    pass

def standorteButtonGeklickt(b):
    global zaehlerDaten
    global zaehlerStandorte
    if((zaehlerDaten + zaehlerStandorte) % 2 == 0):
        IPython.display.clear_output(wait = False)
        display(HBox([buttonDokuDaten,buttonStandort]))
        display(IPython.display.Markdown(standorteDaten))
        zaehlerDaten = zaehlerDaten + 1
        pass
    else:
        if(zaehlerDaten % 2 == 0):
            IPython.display.clear_output(wait = False)
            display(HBox([buttonDokuDaten,buttonStandort]))
            display(IPython.display.Markdown(standorteDaten))
            zaehlerStandorte = zaehlerStandorte + 1
            zaehlerDaten = zaehlerDaten + 1
            pass
        else:
            IPython.display.clear_output(wait = False)
            display(HBox([buttonDokuDaten,buttonStandort]))
            zaehlerDaten = zaehlerDaten + 1
            pass
        pass
    pass
    
buttonStandort.on_click(standorteButtonGeklickt)
display(HBox([buttonDokuDaten,buttonStandort]))
```

## Tagesablauf auf einer Karte einzeichnen

Mit Hilfe der Heatmap konntet ihr euch bereits einen sehr guten Überblick über den Tag verschaffen. Allerdings ist eine Heatmap nicht sehr detailliert und man kann auch nicht erkennen, wann eine Person einen bestimmten Ort besucht hat. Aus diesem Grund geht es nun darum, die Tagesabläufe der Personen auf der Karte einzuzeichnen, damit detaillierte Informationen zugrunde liegen. Dazu wird ebenfalls eine __Karte__ verwendet. 

```python
#Hier wird eine Karte mit dem Mittelpunkt an den Koordinaten (51.7189205,8.7575093) erstellt
meineKarte = Karte((51.7189205,8.7575093))

#Hier werden alle Standorte des Datensatzes auf die Karte gezeichnet
meineKarte.zeichneStandorteEin(standortdaten, 'blue')

#Hier werden alle Standorte zeitlich sortiert miteinander verbunden
meineKarte.verbindeAlleStandorte(standortdaten)
```

## Personen zuordnen
